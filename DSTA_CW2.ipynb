### CW1 Phase 2

#### Background

The dataset is from The World Happiness Report which is a publication of the UN Sustainable Development Solutions Network.  The report gets its data mainly from the Gallup World Poll although many other sources contribute as well.  The data collected from the Gallup poll between the year 2005-2019, this is because for the countries which had the most recent data missing, the next most recent data is taken.  The report choses a particular focus each year, the 2020 report focuses on the natural, urban, and social environment.  There is one Kaggle challenge on the dataset, which is to simply compare the countries by happiness and other human metrics.  This challenge does not ask of anything specific and is very open ended.

#### Variables

The dataset includes 153 rows (observations, data points) and 20 columns (attributes).  Among these attributes 18 are numeric attributes and 2 are categorical (country name and reginal indicator).   The 153 observations are the countries which are ranked in the index.  

Categorical Attributes:

1.	**Country Name** – the name of the country, e.g. “Japan”
2.	**Regional Indicator** – the region the country is geographically located in, e.g. “East Asia”


Numerical Attributes:

1.	**Ladder Score** – the happiness score of the country, a value between ten and zero. The score is the national average of surveys done in each country asking the citizens how they would score their lives, 0 being the worst possible life, 10 being the best.
2.	**Standard error of the ladder score** – the standard error of the ladder score
3.	**Upper whisker** – the ladder score at the upper quartile (Q3) of the range
4.	**Lower whisker** – the ladder score at the lower quartile (Q1) of the range
5.	**Logged GDP per capita** – this takes the natural log (log base e) of each country’s GDP per capita (purchasing power parity) based on 2011 international dollars.
6.	**Social support** – is the national average of the binary response (0 or 1) to the question “If you were in trouble, do you have relatives or friends you can count on to help you whenever you need them, or not?”.  The value is between 1 and 0.
7.	**Healthy life expectancy** – this data is how many years a citizen of that country is expected to live in a healthy manner.  Taken from WHO’s Health Observatory data repository.
8.	**Freedom to make life choices** – this is the national average responses to the question “Are you satisfied or dissatisfied with your freedom to choose what you do with your life?”.
9.	**Generosity** – is the residual of regressing the national average of GWP responses to the question “Have you donated money to a charity in the past month?” on GDP per capita.  It can be positive or negative.
10.	**Perceptions of corruption** – these are the national average of responses to two binary questions on corruption from GWP (Gallup World Poll).  The questions are “Is corruption widespread throughout the government or not?” and “Is corruption widespread within businesses or not?”.
11.	**Ladder score in Dystopia** – this value is a constant for all observations.  It is 1.97.  It represents the worst possible life, and it is used to explain a small part (1.97) of the ladder score.  In other words, it serves as a benchmark to see the contributions of other factors to the happiness ladder score.
12.	**Explained by: Log GDP per capita** – part of the ladder score explained by Log GDP per capita.
13.	**Explained by: Social Support** – part of the ladder score explained by Social Support.
14.	**Explained by: Healthy life expectancy** – part of the ladder score explained by Healthy life expectancy.
15.	**Explained by: Freedom to make life choices** – part of the ladder score explained by Freedom to make life choices.
16.	**Explained by: Generosity** – part of the ladder score explained by Generosity.
17.	**Explained by: Perceptions of corruption** – part of the ladder score explained by Perceptions of Corruption.
18.	**Dystopia + residual** – this value is the sum of 2 values.  First is the score from dystopia which is 1.97.  The second is the average residual for each nation.  The residual is different for each country and it represents the part of the national ladder score that can not be explained by the model constructed by this index.

*Although not directly quoted, much of the info on attributes is from Statistical Appendix for Chapter 2 of World Happiness Report 2020.

### CW2 Phase 1

To recap CW1, this analysis focuses on World Happiness Report 2020 from the Kaggle World Happiness Report up to 2020 dataset.  The dataset contains 153 observations and 20 attributes, among these 20 attributes the most important 9 were : “Country name”, “Regional indicator”, “Ladder Score”, “Logged GDP per Capita”, “Social Support”, “Healthy life expectancy”, “Freedom to make life choices”, “Generosity” and “Perceptions of Corruption”

Among these 9 the most important attributes, I need to determine a 3 predictors and 1 predicted dimension.  The predicted dimension should can be many things, for example the perceptions of corruption of a country could be predicted based on other variables and so on.  But here the name of the dataset implies that our aim should be to estimate the overall happiness so the predicted dimension should be the "Ladder Score".  I should remind that the "Ladder score" from the Dataset is not calculated from other attributes but the score is the result of a poll done by Gallup World Poll, in which people self report their own assesments of their well being.

For this exercise I would like to have a binary classification task, so I will be dividing the countries into 2 categories, happy and unhappy ones.  Therefore I will be introducing a new column named "Happiness".  This coulmn will have "happy" for countries with above average happiness scores and "unhappy" for those below average scores.  This "Happiness" will be my predicted dimension.  It may be converted to 1 or 0 if desired.

When it comes to picking our 3 predictors we have 6 candidates: “Logged GDP per Capita”, “Social Support”, “Healthy life expectancy”, “Freedom to make life choices”, “Generosity” and “Perceptions of Corruption”.  

According to the World Happiness Report 2020 page 18, the 3 attributes that make the largest parts of the average country's Ladder Score are “Social Support” (33%), “Logged GDP per Capita” (25%) and “Healthy life expectancy” (20%).  So these 3 will be my predictors.

```python
import csv
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# importing the dataset using pandas
Whr2020 = pd.read_csv('2020.csv')

# calculation the mean Ladder Score
mean_country = Whr2020["Ladder score"].mean()

# adding a new dimension "Happiness" which is binary
Whr2020.loc[Whr2020["Ladder score"] > mean_country, "Happiness"] = "happy"
Whr2020.loc[Whr2020["Ladder score"] < mean_country, "Happiness"] = "unhappy"

# generating scatter plot matrix for predictor variables,
# the predicted variable is indicated with color
sns.pairplot(Whr2020[["Happiness", "Social support", 
                      "Logged GDP per capita", 
                      "Healthy life expectancy"]], hue="Happiness")
```

The plots above are generated using seaborn package and the data is imported with pandas.  Instead matplotlib and numpy could have been used from session 1c but I used seaborn and pandas to keep the code much shorter and simpler.

The three dimensions we are looking at, when taken together, can become a decent predictor for the "Happiness" dimension.  As we can see from the plots, none of predictors alone provide enough separation alone, but together they can make a reasonable prediction.

### CW2 Phase 2

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# creating a dataset with only the relevant dimensions
Whr2020_reduced = Whr2020[["Happiness", "Social support", 
                           "Logged GDP per capita", 
                      "Healthy life expectancy"]]
```

```python
Whr2020_reduced.describe()
```

The table above shows that the three predictor dimensions have very different ranges and means.  For example Social support has an range of 0.65, while the range for Healthy life expectancy is 31.60.  This may be a problem because in PCA what we are looking for is the variance in each attribute.  So if values aren't scaled, attributes that have much greater ranges like "Healthy life expectancy" will dominate the attributes with much smaller ranges like "Social support" in terms of distance.

```python
# making the mean of all predictors 0 and standard deviations 1.

Whr2020_predictors = Whr2020_reduced.drop("Happiness", axis = 1)

for column in Whr2020_predictors.columns:
    Whr2020_predictors[column] = (Whr2020_predictors[column] - 
                                  Whr2020_predictors[column].mean())
    / Whr2020_predictors[column].std()

# converting the "Hapiness" variable to 1(happy) and 0(unhappy).
Whr2020_predictors["Happiness"] = Whr2020_reduced["Happiness"]

Whr2020_predictors.loc[Whr2020_predictors["Happiness"] == "happy",
                       "Happiness"] = 1
Whr2020_predictors.loc[Whr2020_predictors["Happiness"] == "unhappy",
                       "Happiness"] = 0

Whr2020_predictors_h = Whr2020_predictors
Whr2020_predictors = Whr2020_predictors.drop("Happiness", axis = 1)

# here it got a bit confusing, I ended up with 2 datasets, 
# 1 with only the 3 predictors named Whr2020_predictors,
# and another dataset with both the predictors and 
# the predicted named Whr2020_predictors_h,
# I will only be extracting the y (target) variable 
# from Whr2020_predictors_h dataset in the future
# the dataset used for PCA won't have the target variable in it as a dimension
```

The first step of the PCA will be to substract the means of each variable from the values in that variable so all the means in my new dataset will be 0.  This is because in PCA means don't have an importance, what is important is the variance.  The second step is to scale the data to have an std of 1.  The reason for this was explained above.  I did these 2 with the code above.

```python
# 2d PCA

y = Whr2020_predictors_h["Happiness"]

target_names = ["unhappy", "happy"]

plt.figure()

pca = PCA(n_components=2)
X_r = pca.fit(Whr2020_predictors).transform(Whr2020_predictors)

colors = ['navy', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('First two PCA directions')

print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
# 3d PCA

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

y = Whr2020_predictors_h["Happiness"]

X_reduced = PCA(n_components=3).fit_transform(Whr2020_predictors)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set3, edgecolor='k', s=50)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])

ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])

ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()

pca = PCA(n_components=3)
X_r = pca.fit(Whr2020_predictors).transform(Whr2020_predictors)

# Percentage of variance explained for each components

print('explained variance ratio (three components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
# Screeplot
pca = PCA(n_components=3)
X_r = pca.fit(Whr2020_predictors).transform(Whr2020_predictors)

plt.title("Screeplot")
plt.plot(pca.explained_variance_ratio_, linestyle = "--", marker = "o")
```

In the screeplot above the first dot represents the first PCA, the second dot the second PCA and the third dot the third PCA.

### CW2 Phase 3

#### Description of results

The screeplot and the values obtained show that the first PCA is responsible for 86% of the variance while the second PCA is responsible for 9% and the third PCA is responsible for 5% of the variance.  This indicates that the data can be represented with only 2 PCAs while keeping 95% of the variance.  I did two PCAs in phase 2, one is 2D with the first two PCAs and the other is 3D with the first three PCAs.  

The separation of two groups based on the target variable is discernable in both cases.  Both cases resulted in similar levels of separation which is consistant with the finding that the first two PCAs represent 95% of the variance.  Since having a third PCA adds such a small level of further variance, having 2 PCAs can be preferred.

#### Motivation

The reason I picked three predictor dimensions, "Social support", "Logged GDP per capita" and "Healthy life expectancy" is that the WHR2020 states that these 3 predictors are responsible for most of the Hapiness (measured in the Ladder Score).  According to WHR2020 page 18, Social support is responsible for 33%, GDP per capita 25%, healthy life expectancy 20%, freedom 13%, generosity 5% and corruption 4% of the average country's Ladder score.  So I picked the three that made up the largest part of the Ladder score, which I based my happiness classifier on.

Here I made the assumption that the six variables are independent of one another, this assumption is probably wrong.  There can be many linkages between the variables, like healthy life expectancy can be dependent of a country's health system, which could have more resources if the gdp per capita is higher.  Some of these possibilities are also entertained in the WHR2020 pg 18-19, but ultimately these variables and methods are decided as the best possible practice so I treated the variables as inpedendent.

The other alternative variables are the "Freedom to make life choices", "Generosity" and "Perceptions of corruption".  Using any of these variables instead of the ones I picked will most likely produce worse results.  To check this I plotted all the 6 posible predictors below in scatter and line plots.

```python
# creating a dataset with all 6 predictor dimensions

Whr2020_all = Whr2020[["Happiness", "Social support", "Logged GDP per capita", 
                      "Healthy life expectancy", "Freedom to make life choices",
                      "Generosity", "Perceptions of corruption"]]

# creating a dataset with all 6 predictor dimensions

sns.pairplot(Whr2020_all[["Happiness", "Social support", 
                          "Logged GDP per capita", 
                      "Healthy life expectancy", "Freedom to make life choices",
                      "Generosity", "Perceptions of corruption"]], hue="Happiness")
```

As one can see, the bottom three variables (the ones I didn't pick) provide much worse separation for the target variable than the top three (the ones I picked).

An alternative approach could have been to keep all 6 variables in the dataset and run the PCA again.  I did that below to see the results.

```python
# making the mean of all predictors 0 and standard deviations 1.

Whr2020_6predictors = Whr2020_all.drop("Happiness", axis = 1)

for column in Whr2020_6predictors.columns:
    Whr2020_6predictors[column] = (Whr2020_6predictors[column] - 
                                   Whr2020_6predictors[column].mean()) 
    / Whr2020_6predictors[column].std()

Whr2020_6predictors["Happiness"] = Whr2020_all["Happiness"]

Whr2020_6predictors.loc[Whr2020_6predictors["Happiness"] == "happy",
                        "Happiness"] = 1
Whr2020_6predictors.loc[Whr2020_6predictors["Happiness"] == "unhappy", 
                        "Happiness"] = 0

Whr2020_6predictors_h = Whr2020_6predictors
Whr2020_6predictors = Whr2020_6predictors.drop("Happiness", axis = 1)
```

```python
# Screeplot

pca = PCA(n_components=6)
X_r = pca.fit(Whr2020_6predictors).transform(Whr2020_6predictors)

plt.title("Screeplot")
plt.plot(pca.explained_variance_ratio_, linestyle = "--", marker = "o")
```

```python
# 2d PCA

y = Whr2020_6predictors_h["Happiness"]

target_names = ["unhappy", "happy"]

plt.figure()

pca = PCA(n_components=2)
X_r = pca.fit(Whr2020_6predictors).transform(Whr2020_6predictors)

colors = ['navy', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('First two PCA directions')

print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
# 3d PCA

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)

y = Whr2020_6predictors_h["Happiness"]

X_reduced = PCA(n_components=3).fit_transform(Whr2020_6predictors)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set3, edgecolor='k', s=50)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])

ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])

ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()

pca = PCA(n_components=3)
X_r = pca.fit(Whr2020_6predictors).transform(Whr2020_6predictors)

# Percentage of variance explained for each components

print('explained variance ratio (three components): %s'
      % str(pca.explained_variance_ratio_))
```

The result of the PCA with all 6 predictors does not provide better separation than the PCA done with the most important 3 predictors.  The Screeplot for this case reveals that using only the first 2 PCAs would lose too much variance in this case and the poor result seen in the 2D PCA plot confirms this.  In this case using the best three PCAs can produce a decent separation, this is seen in the 3D plot above.  But overall the results are underwhelming compared to the PCA done only with the most important 3 variables.
